```python
import pandas as pd
```

```python
#scrape data
string = "https://www.pro-football-reference.com/years/{}/fantasy.htm"
years = ["2017","2018","2019","2020","2021"]
data_dict = {}
for year in years:
    url = string.format(year)
    data = pd.read_html(url, header = 1)
    data_dict[year] = pd.DataFrame(data[0])
    
data_dict['2021'].dtypes
```

```python
#clean NaN
for x,y in data_dict.items():
    data_dict[x] = data_dict[x].drop(data_dict[x][data_dict[x].Age == 'Age'].index)
    data_dict[x] = data_dict[x].dropna(subset = ['FantPos'])
    data_dict[x] = data_dict[x].fillna(0)
    print(data_dict[x].shape)
```
